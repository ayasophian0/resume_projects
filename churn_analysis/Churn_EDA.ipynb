# #Determines
In this project you have HR data of a company. A study is requested from you to predict which employee will churn by using this data.

The HR dataset has 14,999 samples. In the given dataset, you have two types of employee one who stayed and another who left the company.

You can describe 10 attributes in detail as:
- ***satisfaction_level:*** It is employee satisfaction point, which ranges from 0-1.
- ***last_evaluation:*** It is evaluated performance by the employer, which also ranges from 0-1.
- ***number_projects:*** How many of projects assigned to an employee?
- ***average_monthly_hours:*** How many hours in averega an employee worked in a month?
- **time_spent_company:** time_spent_company means employee experience. The number of years spent by an employee in the company.
- ***work_accident:*** Whether an employee has had a work accident or not.
- ***promotion_last_5years:*** Whether an employee has had a promotion in the last 5 years or not.
- ***Departments:*** Employee's working department/division.
- ***Salary:*** Salary level of the employee such as low, medium and high.
- ***left:*** Whether the employee has left the company or not.

## 1. Exploratory Data Analysis

### 1.1. Apply Data Integrity Checks
* link: https://docs.deepchecks.com/stable/user-guide/tabular/auto_quickstarts/plot_quick_data_integrity.html

```python
df.columns
```

```python
df.columns = ['satisfaction_level', 'last_evaluation', 'number_project',
       'average_monthly_hours', 'time_spend_company', 'work_accident', 'left',
       'promotion_last_5years', 'departments', 'salary']
```

```python
from deepchecks.tabular import Dataset
from deepchecks.tabular.suites import data_integrity

cats = ['departments', 'salary']

ds = Dataset(df, cat_features=cats, label='left')

suite = data_integrity()
result = suite.run(ds)
```

```python
result.show()
```

```python
df.drop_duplicates(inplace=True)
```

## 2. Data Visualization

```python
df.select_dtypes(include=np.number).iloc[:, :-3].plot(kind='box', subplots=True, figsize=(10,5), 
                                         title='Overall Distribution of Non-Binary Numeric Features')
plt.tight_layout()
plt.show()
```

```python
df.satisfaction_level.plot(kind='hist', xlabel='Satisfaction Level');
```

```python
df.last_evaluation.plot(kind='hist', xlabel='Last Evaluation');
```

```python
df.time_spend_company.plot(kind='hist', xlabel='Time Spent in Company');
```

```python
df[df.satisfaction_level < 0.21].left.value_counts().plot(kind='bar', title="Low Satisfaction Level's Effect On Leaving");
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,5))
columns = ['work_accident', 'left', 'promotion_last_5years']

for ax, col in zip(axes, columns):
    sns.countplot(data=df, x=col, ax=ax)
    for container in ax.containers:
        ax.bar_label(container)

fig.suptitle('Distribution of Boolean Features', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df[df.promotion_last_5years == 1].satisfaction_level.plot(kind='hist', xlabel='Satisfaction Level With Promotion');
```

```python
df[df.promotion_last_5years == 0].satisfaction_level.plot(kind='hist', xlabel='Satisfaction Level Without Promotion');
```

```python
ax = sns.countplot(data=df, x='salary', hue='salary')
plt.title('Distribution of Salary Bins')
for container in ax.containers:
    plt.bar_label(container);
```

```python
ax = df.groupby('salary')['average_monthly_hours'].mean().plot(kind='bar', title='Avg Monthly Hours by Salary', 
                                                       ylabel='Avg Monthly Hours')

for container in ax.containers:
    plt.bar_label(container, fmt='%.2f');
```
